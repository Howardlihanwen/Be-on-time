```python
import os
import pandas as pd
import numpy as np
import re

# 指定目錄檔路徑
path = 'C:/Users/student/期末project/02_fp_data_weather2'
# 取得目錄下所有的csv檔案的路徑
file_paths = [f for f in os.listdir(path) if f.endswith('.csv')]
```

```python
def stationPressure():
    # 將資料轉形成字串
    df['station_pressure'] = df['station_pressure'].astype(str)
    df['station_pressure'] = df['station_pressure'].str.extract('(\d+\.\d{2})', expand=True)
    # 將資料轉換成浮點數
    df['station_pressure'] = pd.to_numeric(df['station_pressure'], errors='coerce')
    # 插補缺失值
    df['station_pressure'] = df['station_pressure'].interpolate(limit_direction='both', limit=300)
    df['station_pressure'] = df['station_pressure'].fillna(df['station_pressure'].mean())
```

```python
def RSTY():
    # 先處理欄位型態轉制成float型態,透過隱含函數lambda
    df['relative_humidity']=  df['relative_humidity'].apply(lambda x : eval(x.replace(x,'0')) if x =='*' else float(x) )
    df['wind_speed']=  df['wind_speed'].apply(lambda x : x.replace(x,'0') if x =='*' else x )
    df['wind_speed']= df['wind_speed'].apply(lambda x: eval(x.strip('s')) if type(x)==str else x)
    df['visibility']=  df['visibility'].apply(lambda x : x.replace(x,'0') if x =='*' else x )
    df['visibility']= df['visibility'].apply(lambda x: eval(x.strip('Vs')) if type(x)==str else x)
    
    # 將亂數及亂碼修正
    # 將(空缺nan)float 換成字串 
    df['wind_dict']=  df['wind_dict'].apply(lambda x : str(x) if type(x)!=str else x )
    # 將'VRB' 換成 False 且以字串 0 代表
    df['wind_dict']=  df['wind_dict'].apply(lambda x : x.replace(x,'1') if x =='VRB' else x )
    # 將'nan'字串 換成 False 且以字串 0 代表
    df['wind_dict']=  df['wind_dict'].apply(lambda x : x.replace(x,'1') if x=='nan' else x )
    # 將'數值'字串 換成 True 且以字串 1 代表
    df['wind_dict']=  df['wind_dict'].apply(lambda x : x.replace(x,'0') if x!=('nan','VRB','1') else x )
    # 將'0' '1'字串 轉成數值
    df['wind_dict']=  df['wind_dict'].apply(lambda x : eval(x) )
    
    # 插補缺失值
    df[['relative_humidity']] = df[['relative_humidity']].interpolate(method='nearest' , limit_direction = 'both', limit = 300 )
    df[['relative_humidity']] = df[['relative_humidity']].fillna(df['relative_humidity'].mean())
    df[['wind_speed']] = df[['wind_speed']].interpolate(method='nearest' ,limit_direction = 'both', limit = 1000)
    df[['wind_speed']] = df[['wind_speed']].fillna(df['wind_speed'].mean())
    df[['visibility']] = df[['visibility']].interpolate(method='nearest' ,limit_direction = 'both', limit = 300)
    df[['visibility']] = df[['visibility']].fillna(df['visibility'].mean())
    
```

```python
def LMNtemp():
    # 將列轉換為字符串類型
    df['dry_bulb_temp'] = df['dry_bulb_temp'].astype(str)
    df['wet_bulb_temp'] = df['wet_bulb_temp'].astype(str)
    df['dew_point_temp'] = df['dew_point_temp'].astype(str)
    
    # 使用正則表達式將非數字字元替換為空字串，然後轉換為浮點數
    df['dry_bulb_temp'] = df['dry_bulb_temp'].apply(lambda x: re.sub('[^0-9.]', '', x))
    df['dry_bulb_temp'] = pd.to_numeric(df['dry_bulb_temp'], errors='coerce')
    df['wet_bulb_temp'] = df['wet_bulb_temp'].apply(lambda x: re.sub('[^0-9.]', '', x))
    df['wet_bulb_temp'] = pd.to_numeric(df['wet_bulb_temp'], errors='coerce')
    df['dew_point_temp'] = df['dew_point_temp'].apply(lambda x: re.sub('[^0-9.]', '', x))
    df['dew_point_temp'] = pd.to_numeric(df['dew_point_temp'], errors='coerce')
    
    # 使用interpolate()方法進行線性插值
    df['dry_bulb_temp'] = df['dry_bulb_temp'].interpolate(method='nearest', limit_direction = 'both', limit = 300)
    df['dry_bulb_temp'] = df['dry_bulb_temp'].fillna(df['dry_bulb_temp'].mean())
    df['wet_bulb_temp'] = df['wet_bulb_temp'].interpolate(method='nearest', limit_direction = 'both', limit = 300)
    df['wet_bulb_temp'] = df['wet_bulb_temp'].fillna(df['wet_bulb_temp'].mean())
    df['dew_point_temp'] = df['dew_point_temp'].interpolate(method='nearest', limit_direction = 'both', limit = 300)
    df['dew_point_temp'] = df['dew_point_temp'].fillna(df['dew_point_temp'].mean())



    
```

```python
def cleanSnow():
    # 清整「snow_fall」欄位: 轉換 'T' 值並替換成 0.05
    df['snow_fall'] = df['snow_fall'].replace('T', 0.05)
    # 清整「snow_depth」欄位:用正則表達式取出T值轉換為數值0.5
    df['snow_depth'] = df['snow_depth'].replace('T', 0.5)
    
    # 將資料轉換成浮點數
    df['snow_fall'] = pd.to_numeric(df['snow_fall'], errors='coerce')
    df['snow_depth'] = pd.to_numeric(df['snow_depth'], errors='coerce')
    
    # 清整「snow_fall」欄位:使用interpolate()方法進行線性插值
    #df['snow_fall'] = df['snow_fall'].interpolate(method='nearest', limit_direction = 'both', limit = 300)
    df['snow_fall'] = df['snow_fall'].fillna(method='ffill')
    # 清整「snow_depth」欄位:使用interpolate()方法進行線性插值
    #df['snow_depth'] = df['snow_depth'].interpolate(method='nearest', limit_direction = 'both', limit = 300)
    df['snow_depth'] = df['snow_depth'].fillna(method='ffill')
    
    # 「snow_fall」「snow_depth」欄位剩餘NAN值補上0
    df[["snow_fall","snow_depth"]] = df[["snow_fall","snow_depth"]].fillna(0)
    
```

```python
def cleanPrecip():
    
    # 清整 'precipitation' 欄位: 轉換 'T' 值並替換成 0.005
    df['precipitation'] = df['precipitation'].replace('T', 0.005)
    
    # 將資料轉換成浮點數
    df['precipitation'] = pd.to_numeric(df['precipitation'], errors='coerce')
    
    # 清整 'precipitation' 欄位: 使用interpolate()方法進行線性插值
    df['precipitation'] = df['precipitation'].interpolate(method='nearest', limit_direction='both', limit=300)
    df['precipitation'] = df['precipitation'].fillna(df['precipitation'].mean())
    
    # 欄位剩餘NAN值補上0
    df['precipitation'] = df['precipitation'].fillna(0)
    
```

```python
# 讀取所有檔案
for file_path in file_paths:
    # load csv檔案為 DataFrame 對象
    file_path = os.path.join(path, file_path)
    df = pd.read_csv(file_path, encoding="utf-8")
    # 執行資料處理
    stationPressure()
    RSTY()
    LMNtemp()
    cleanSnow()
    cleanPrecip()
    # 將處理後的 DataFrame 存成 csv 檔案到新的路徑
    new_file_path = os.path.join('C:/Users/student/Desktop/weatherdata', os.path.basename(file_path))
    df.to_csv(new_file_path, index=False)
```

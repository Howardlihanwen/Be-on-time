```python
'''目標：計算 data 資料總筆數'''
import os
import pandas as pd

# 原資料夾路徑
folder_path = 'C:/Users/o1nod/OneDrive/桌面/01_raw_data_weather/'

# 取得資料夾中所有 CSV 檔案的檔名
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# 計數器
countdata = 0

# 逐一處理每一個 CSV 檔案
for file in csv_files:
    # 讀取 CSV 檔案
    df = pd.read_csv(os.path.join(folder_path, file), dtype=object, index_col=False)
    print(f"檔案 {file} 有 {len(df)} 筆資料")
    countdata += len(df)

print(countdata)
```

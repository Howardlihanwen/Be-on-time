```python
'''第三階段爬蟲'''
'''目標：下載 NOAA 天氣觀測站點 2017~2022 年天氣資料的.csv order'''
'''匯入套件'''
import os, json

# 操作 browser 的 驅動程式
from selenium import webdriver

# 負責開啟和關閉 Chrome 的套件
from selenium.webdriver.chrome.service import Service

# 自動下載 Chrome Driver 的套件
from webdriver_manager.chrome import ChromeDriverManager

# 強制停止/強制等待 (程式執行期間休息一下)
from time import sleep

# 隨機取得 User-Agent
from fake_useragent import UserAgent
ua = UserAgent()

# 啟動瀏覽器工具的選項
my_options = webdriver.ChromeOptions()
# my_options.add_argument("--headless")                # 不開啟實體瀏覽器背景執行
my_options.add_argument("--start-maximized")         # 最大化視窗
my_options.add_argument("--incognito")               # 開啟無痕模式
my_options.add_argument("--disable-popup-blocking")  # 禁用彈出攔截
my_options.add_argument("--disable-notifications")   # 取消通知
my_options.add_argument(f'--user-agent={ua.random}') # (Optional)加入 User-Agent


# 使用 Chrome 的 WebDriver
driver = webdriver.Chrome(
    options = my_options,
    service = Service(ChromeDriverManager().install())
)
```

```python
# 設定要讀入的json檔存放位置
path_to_json = 'C:/order_json/'

# 將資料夾內所有.json檔一次讀進來
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
```

```python
# 建立空list
listResult = []

# 將所有json append進同一個list
for i in json_files:
    with open(f'{path_to_json}{i}', 'r', encoding='utf-8') as file:
        data = file.read()
        listResult += json.loads(data)
```

```python
len(listResult)
```

```python
for i in range(len(listResult)):
    driver.get(f'https://www.ncei.noaa.gov/orders/cdo/{listResult[i]}.csv')
    print(f'第{i}筆, OK')
```
